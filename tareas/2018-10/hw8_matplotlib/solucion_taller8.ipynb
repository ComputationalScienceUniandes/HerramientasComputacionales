## Punto 2
definición de función

```python
def V(w,L):
    num=10.0*w*L
    enraiz=2.0*L**4*w**4 - 2.0*L**2*w**2 + 1
    den=sqrt(enraiz)
    return num/den
```

evalúo para verificar que esté bien implementada

```python
V(1,1)
```

## Punto 3
Creo vector logarítmico de 10 hasta 100.000

```python
vecw=logspace(1,5,100)
```

## Punto 4
Evalúo y grafico función para L=1mH

```python
Lp4=V(vecw,0.001)
```

```python
plot(vecw,Lp4)
show()
```

## Punto 5
Las 3 curvas, para L=1mH,L=2mH, y L=5mH.

```python
L1=V(vecw,0.001)
L2=V(vecw,0.002)
L5=V(vecw,0.005)
plot(vecw,L1,vecw,L2,vecw,L5)
title('circuito')
xlabel('frecuencia')
ylabel('voltaje')
legend(['L=1mH','L=2mH','L=5mH'])
show()
```

## Punto 6
Importar los datos. Tener en cuenta que hay una fila de encabezados a ignorar.

```python
datosRLC = np.loadtxt("rlc.txt",skiprows=1)
```

Se grafican los datos como puntos. En este caso la primera columna es eje x (frecuencia) y la segunda columna es eje y (L).

```python
L1=V(vecw,0.001)
L2=V(vecw,0.002)
L3=V(vecw,0.005)
plot(vecw,L1,vecw,L2,vecw,L3)
plot(datosRLC[:,0],datosRLC[:,1],marker='.',markersize=5,linewidth=0)
title('circuito')
xlabel('frecuencia')
ylabel('voltaje')
legend(['L=1mH','L=2mH','L=5mH','datos'])
show()
```

## Punto 7
En escala logaritmica en x, uso semilogx

```python
L1=V(vecw,0.001)
L2=V(vecw,0.002)
L3=V(vecw,0.005)
semilogx(vecw,L1,vecw,L2,vecw,L3)
semilogx(datosRLC[:,0],datosRLC[:,1],marker='.',markersize=5,linewidth=0)
title('circuito')
xlabel('frecuencia')
ylabel('voltaje')
legend(['L=1mH','L=2mH','L=5mH','datos'])
show()
```
