```python
# La siguiente función recibe los parámetros de la simulación e
# inicializa
def inicializar(numCarros0,roadLen0,vmax0,p0):
    global vel, pos, numCarros, roadLen, vmax, p
    numCarros=numCarros0
    roadLen=roadLen0
    vmax=vmax0
    p=p0
    ## Inicializar la via
    via=concatenate((1+zeros(numCarros,int),zeros(roadLen-numCarros,int)))
    random.shuffle(via)
    ## Inicializar las velocidades
    vel=zeros(numCarros,int)
    ## Inicializar las posiciones
    pos=where(via)[0]
    ## Inicializar las distancias de separación
    dists()

# La siguiente función actualiza las distancias 
# de separación entre los carros.
    
def dists():
    global pos, roadLen, seps
    seps=mod(roll(pos,-1)-pos,roadLen)
    
# La siguiente función implemente el primer paso
def nagel1():
    global vmax, vel
    vel=where(vel<vmax,vel+1,vel)

# La siguiente función implementa el segundo paso
def nagel2():
    global vel, pos
    # Actualizar las distancias de separación
    dists()
    vel=where(vel>=seps,seps-1,vel)
    
# La siguiente función implementa el tercer paso
def nagel3():
    global vel, numCarros
    # Encontrar los carros con velocidad positiva
    movin=where(vel>0)[0]
    # Lanzar una moneda para cada carro con velocidad positiva
    monedas=random.random(len(movin))
    # Elegir los carros a los que van a reducirse las velocidades 
    magicCars=movin[where(monedas<p)]
    magicVels=vel[magicCars]
    vel[magicCars]=magicVels-1
    
# Esta función implementa los tres pasos y actualiza las posiciones
def nagel():
    global pos, vel
    nagel1()
    nagel2()
    nagel3()
    pos=mod(pos+vel,roadLen)
    
# Esta función convierte la lista de posiciones en el el estado de la vía
def theroad(carpositions):
    road=zeros(roadLen)
    road[carpositions]=road+1
    return road

# Esta función calcula la distancia total recorrida
def totaldistance():
    totdist=0
    for i in range(1000):
        nagel()
        totdist+=sum(vel)
    return totdist
```

```python
#%%timeit
distancias=[]
listCarros=arange(10,250,10)
for i in listCarros:
    inicializar(i,1000,5,0.25)
    distancias.append(totaldistance())
```
