```python
def simulate(k):
    via = zeros(1000)
    pos = zeros(k)
    vel = zeros(k)
    vmax = 5
    dtotal = 0
    mat = [None for i in range(1000)]
    
    # El estado inicial de la via, con k carros es generado aleatoreamente. Se guardan las posiciones de los carros
    # en el arreglo pos.
    for i in range(k):
        r = randint(0,1000)
        while(via[r] == 1):
            r = randint(0,1000)
        via[r] = 1
        pos[i] = r
        
    # Ordenamos pos de menor a mayor. Y a la matriz con la simulacion entera le ponemos el estado inicial de la via.
    pos = sort(pos)
    mat[0] = via
    
    # Ahora vamos a hacer cada uno de los pasos del modelo. Ademas, lo agregamos a la matriz con toda la simulacion.
    for j in range(999):
        npos = zeros(k)
        nvia = zeros(1000)
        for i in range(k):
            if i == k-1:
                vel[i] = min(vel[i]+1,vmax)
                d = 1000 + pos[0]-pos[i]
                vel[i] = min(vel[i],d-1)
                if (randint(0,3)==0):
                    vel[i] = max(0,vel[i]-1)
                npos[i] = pos[i] + vel[i]
                if npos[i] >= 1000:
                    npos[i] = npos[i] - 1000
            else:
                vel[i] = min(vel[i]+1,vmax)
                d = pos[i+1]-pos[i]
                vel[i] = min(vel[i],d-1)
                if (randint(0,3)==0):
                    vel[i] = max(0,vel[i]-1)
                npos[i] = pos[i] + vel[i]
        dtotal += sum(vel)
        if npos[k-1] < npos[k-2]:
            npos = sort(npos)
            rep = vel[k-1]
            for i in range(k-1):
                vel[i+1] = vel[i]
            vel[0] = rep
        for i in range(k):
            pos[i] = npos[i]
            nvia[npos[i]] = 1
        mat[j+1] = nvia
        
    return mat, dtotal
```

```python
k1 = 170
```

```python
simu = simulate(k1)[0]
```

```python
figsize(10,10)
imshow(simu, cmap = 'Greys', interpolation = 'nearest')
savefig('traffic.png')
```
