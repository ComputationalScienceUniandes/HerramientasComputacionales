# Ajustes con Scipy

Primero vamos a graficar el área despejada acumulada en función del tiempo

```python
figure(figsize(15,10))
plot(totalyear,area, 'ko')
ticklabel_format(useOffset=False)
xlim(2006,2015)
xlabel(u'Año')
ylabel(u"Área despejada acumulada / hm^2")
show()
```

Noten que a medida que aumenta el tiempo de desminado, tiende a saturarse el área total despejada. Por lo tanto vamos a usar un modelo logístico para nuestro ajuste:

$$f(t) = \frac{L}{1 + \mathrm e^{-k(t-t_0)}}$$ 

Donde: 
- t_0 = valor en x del punto medio,
- L = valor máximo de la curva
- k L /4 = pendiente de la curva

```python
from scipy.optimize import curve_fit
def logisticmodel(t,t0,k,L):
    return L/(1 + exp(-1*k*(t-t0)))
logisticfit=curve_fit(logisticmodel, totalyear,area,p0=(0,0,0))
figure(figsize=(12,7))

dd=linspace(2006,2015,200)
salidafit=logisticmodel(dd,logisticfit[0][0],logisticfit[0][1],logisticfit[0][2])
plot(totalyear,area,'go',label='datos')
plot(dd,salidafit,'k--',label='fit')
xlim(2006,2015)
ylim(0,250)
xlabel(u'año')
ylabel(u'área despejada acumulada / hm^2')
legend(loc = 'best')
ticklabel_format(useOffset=False)
show()
```

### ¿En qué fecha el área acumulada estimada será igual a 300 $hm^2$? (usar Newton-Raphson)

Derivada de la función logística:

$$\frac{k L \exp(-k (x-x0))}{(1 + \exp(-k (x-x0)))^2}$$

```python
#rescatar los parámetros de logisticfit
tcoeff = logisticfit[0][0]
kcoeff = logisticfit[0][1]
Lcoeff = logisticfit[0][2]

# función ajuste logístico
def ajustelogistico(t,areaacum):
    return (Lcoeff/(1 + exp(-1*kcoeff*(t-tcoeff))) - areaacum)

# función derivada ajuste logístico
def derivada_ajustelogistico(t):
    return kcoeff*Lcoeff*exp(-1*kcoeff*(t-tcoeff))/(1 + exp(-1*kcoeff*(t-tcoeff)))**2


def newtonraphson(inicio,areaacum):
    x=inicio
    for i in range(10): # 100 pasos de n-r
        x -= ajustelogistico(x,areaacum)/derivada_ajustelogistico(x)
    return x
```

```python
ndias = newtonraphson(2020,300)
print ndias
# Area despejada acumulada ~ 500m2
print "Se tendrán 300 hm^2 despejados comienzos del 2018."
```
