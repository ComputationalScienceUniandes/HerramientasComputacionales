```python
def compmeshgrid(remin,remax,immin,immax,ene):
    re=linspace(remin,remax,ene)
    im=linspace(immin,immax,ene)
    exis, ye = meshgrid(re,im,sparse=True)
    return exis - 1j*ye
def fun(z):
    return where(z!=0,z**5+z**2,0)
def dfun(z):
    return where(z!=0,5*z**4+2*z,1)
def newrap(z):
    for i in arange(50):
        z-=fun(z)/dfun(z)
    return z
```

```python
# Baja resolución
midpoint=0
wi=2
z = compmeshgrid(midpoint-wi/2,midpoint+wi/2,-wi/2,wi/2,200)
frac = newrap(z)
figure(figsize=(10,10))
axis("off")
imshow(angle(frac))
show()
savefig("/Users/phoenix/Dropbox/UAndes/Herramientas-Computacionales-Private/Talleres/HW7/fractal_lr.png",dpi=20)
```
