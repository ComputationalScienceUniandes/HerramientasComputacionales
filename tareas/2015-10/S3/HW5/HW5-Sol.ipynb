```python
title='    ______  __________  ______  _   __\n   / __ \ \/ /_  __/ / / / __ \/ | / /\n  / /_/ /\  / / / / /_/ / / / /  |/ / \n / ____/ / / / / / __  / /_/ / /|  /  \n/_/     /_/ /_/ /_/ /_/\____/_/ |_/   \n                                      \n    ____  __    ___   ________ __    _____   ________ __\n   / __ )/ /   /   | / ____/ //_/   / /   | / ____/ //_/\n  / __  / /   / /| |/ /   / ,< __  / / /| |/ /   / ,<   \n / /_/ / /___/ ___ / /___/ /| / /_/ / ___ / /___/ /| |  \n/_____/_____/_/  |_\____/_/ |_\____/_/  |_\____/_/ |_|  \n'

# 1 -> A
# 11 -> J
# 12 -> K
# 13 -> Q
# 14 -> A
# 15 -> 2
# ...

# La siguiente función toma las cartas y las baraja
def barajar():
    for i in range(len(cartas)):
        randy=randint(i,len(cartas)-1)
        cartas[i],cartas[randy]=cartas[randy],cartas[i]
# La siguiente función toma las cartas elige una al azar la entrega y la quita de la baraja
# Con dos jugadores no es posible quedarse sin cartas
def damecarta():
    carta=cartas[0]
    del cartas[0]
    return carta
# La siguiente función toma una lista de cartas y las imprime
def simpleprintcards(cardlist):
    lon=len(cardlist)
    k=0
    for i in cardlist:
        if i%13==1:
            stdout.write("A"),
        elif i%13==11:
            stdout.write("J"),
        elif i%13==12:
            stdout.write("Q"),
        elif i%13==0:
            stdout.write("K"),
        else:
            stdout.write(str(i)),
        if k<lon-1:
            stdout.write(", "),
        k+=1
    print ""

# Imprimir cartas especiales

# Con la ayuda de la fuente cards de figlet 
# más un poco de maquillaje, cada lista contiene
# las seis líneas de cada carta.

nicecartas=[
['.------.','|A.--. |','| (\/) |','| :\/: |','| \'--\'A|','`------\''],
['.------.','|2.--. |','| (\/) |','| :\/: |','| \'--\'2|','`------\''],
['.------.','|3.--. |','| :(): |','| ()() |','| \'--\'3|','`------\''],
['.------.','|4.--. |','| :/\: |','| :\/: |','| \'--\'4|','`------\''],
['.------.','|5.--. |','| :/\: |','| (__) |','| \'--\'5|','`------\''],
['.------.','|6.--. |','| (\/) |','| :\/: |','| \'--\'6|','`------\''],
['.------.','|7.--. |','| :(): |','| ()() |','| \'--\'7|','`------\''],
['.------.','|8.--. |','| :/\: |','| :\/: |','| \'--\'8|','`------\''],
['.------.','|9.--. |','| :/\: |','| (__) |','| \'--\'9|','`------\''],
['.------.','|10--. |','| :/\: |','| :\/: |','| \'--10|','`------\''],
['.------.','|J.--. |','| :(): |','| ()() |','| \'--\'J|','`------\''],
['.------.','|Q.--. |','| (\/) |','| :\/: |','| \'--\'Q|','`------\''],
['.------.','|K.--. |','| :/\: |','| :\/: |','| \'--\'K|','`------\'']
]

def fancyprintcards(cardlist):
    print ""
    for i in range(6):
        for j in cardlist:
            print nicecartas[j-1][i],
        print    

# La siguiente función toma una lista de cartas y calcula el total, 
# si se tienen aces se calcula el mayor posible valor menor o igual a 21 o el menor
def totalizador(cardlist):
    totales=[]
    # r es la cantidad de aces que se toman como 11
    for r in range(4):
        total=0
        k=0
        for carta in cardlist:
            # J, Q o Q
            if carta%13==11 or carta%13==12 or carta%13==0:
                total+=10
            # En caso de encontrar un az
            elif carta%13==1:
                if k<r:
                    total+=11
                else:
                    total+=1
                k+=1
            else:
                total+=carta%13
        totales.append(total)
    # Separar los totales entre los que son menores o iguales a 21 y el resto
    menores=[]
    mayores=[]
    for total in totales:
        if total<=21:
            menores.append(total)
        else:
            mayores.append(total)
    # Si hay algun menor entonces entregar el mayor de ellos
    if len(menores)>0:
        return max(menores)
    # De lo contrario entregar el menor de los mayores, que es igual al mínimo total.
    else:
        return min(mayores)

# La siguiente función decide el ganador en función de los valores
# globales de las variables que contienen los totales del jugador y del computador.
# Si hay un empate el computador gana.
def ganador():
    global totalh
    global totalc
    if totalh<=21 and totalh > totalc and totalc<=21:
        print "¡Ha ganado!"
    else:
        print "¡Ha perdido!"
```
