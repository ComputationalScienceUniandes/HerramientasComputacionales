## Impresora Radical

```python
n=10
dj=1
matr=zeros((n,n),int)
i=0
j=0
for k in range(n*n):
    # Poner el número
    matr[i][j]=k+1
    # Solamente debe cambiar la dirección de movimiento si está en el extremo derecho moviéndose hacia la derecha
    if j==n-1 and dj==1:
        # Esto es para que la columna quede igual, aquí se suma más adelante se resta
        j+=dj
        # Invertir la dirección de movimiento
        dj=-dj
        # Saltar al siguiente renglón
        i+=1
    # Solamente debe cambiar la dirección de movimiento si está en el extremo izquierdo moviéndose hacia la izquierda
    if j==0 and dj==-1:
        # Esto es para que la columna quede igual, aquí se resta más adelante se suma
        j+=dj
        # Invertir la dirección de movimiento
        dj=-dj
        # Saltar al siguiente renglón
        i+=1
    # Efectuar el movimiento en la columna
    j+=dj
# Ahora utilizar imshow    
figure()
imshow(matr,interpolation='None')
show()
```

## Higgs

```python
higgs = genfromtxt("/Users/phoenix/Dropbox/UAndes/Herramientas-Computacionales-Private/Talleres/HW6/higgs.csv",delimiter=",")
# Tomar las columnas para hacer las gráficas
dataenergies = higgs[:,0]
dataevents = higgs[:,1]
# Definir las funciones correspondientes a los ajustes
def nohig(x):
    return 27890 - 526.3 * x +  3.410* x** 2 - 0.007502* x **3
def yeshig(x):
    return nohig(x) + 116 * exp(-(x-125)**2/4.2)
energies = linspace(105,153,100)
nohiggsev = nohig(energies)
yeshiggsev = yeshig(energies)
```

```python
figure(figsize=(8, 8))
rc("font",size=15)
errorbar(dataenergies, dataevents,yerr=35,marker='o',
             linestyle='None',color="black",label="Data")
plot(energies,nohiggsev,'r--',label="$Ev$",linewidth=2)
plot(energies,yeshiggsev,'b',label="$Ev^*$",linewidth=2)
title("CMS $\sqrt{s}=7\,\mathrm{TeV}$, $L=5.1\,\mathrm{fb}^{-1}$ $\sqrt{s}=8\,\mathrm{TeV}$, $L = 5.3\,\mathrm{fb}^{-1}$")
minorticks_on()
tick_params('both', length=10, width=1, which='major')
tick_params('both', length=5, width=1, which='minor')
xlabel("$m_{\gamma\gamma}$(GeV)")
ylabel("S/(S+B) Weighted Events / $1.5$ GeV")
ylim([0,1900])
xlim([105,153])
legend(loc=1,frameon=False)
#savefig("/Users/juan/Dropbox/UAndes/Herramientas-Computacionales-Private/Talleres/HW6/CMS-Higgs.pdf")
show()
```
