```python
# 1. Importar la imagen
foto3 = plt.imread("einstein.jpg")
# 2. Recortar dimensiones
foto1 = foto3[:,:,0]
# Note que ya solo hay un numero por pixel, y no 3 como antes
print(foto1)
```

```python
# Mostrar la imagen
im1 = plt.imshow(foto1, cmap='gray')
plt.colorbar(im1)
```

```python
# Invertir la imagen
im2 = plt.imshow(256-foto1, cmap='gray')
plt.colorbar(im2)
```

```python
# Reflejar la imagen en los dos casos

im3 = plt.imshow(np.fliplr(foto1), cmap='gray')
plt.colorbar(im3)
```

```python
im4 = plt.imshow(np.flipud(foto1), cmap='gray')
plt.colorbar(im4)
```
