```python
x = np.arange(1,9)
exam1 = exams.loc[exams.exam == 1]
exam2 = exams.loc[exams.exam == 2]
y1 = []
y2 = []
for i in range(8):
    d1 = exam1['pt'+str(i+1)].as_matrix()
    d1 = d1[~np.isnan(d1)]
    d2 = exam2['pt'+str(i+1)].as_matrix()
    d2 = d2[~np.isnan(d2)]
    y1.append(sum(d1))
    y2.append(sum(d2))
```
